```python
import pandas as pd
df =pd.read_csv('../data/slack_data.csv', index_col=False).drop(columns=['Unnamed: 0', 'channel', 'subtype'])
df['date_str'] = pd.to_datetime(df['date_str'])
df
```

```python
speaker_info= {}
for i, name in enumerate(df['user_profile'].value_counts().index):
    speaker_info[name] = i

speaker_info_by_id = []
for i, name in enumerate(df['user_profile'].value_counts().index):
    speaker_info_by_id.append({'id': i, 'name': name})

transcript = {
    "name": "nexoya - developer channel",
    "language": "ENGLISH",
    "transcription_backend": "GOOGLE",
    "date": "2022-08-01T19:21:03.901212+00:00",
    "diarization_backend": "AMAZON",
    'speaker_info': speaker_info_by_id,
    'transcript': {
            'type': 'doc',
                'content': [
        {
            'type': 'transcription',
            'content': []
        }
    ]
    }

}

for i in df.iterrows():
    if (i[1]['user_profile'] in speaker_info):
        startTime = i[0] * 100
        endTime = startTime + 100

        speaker_frame = {
            'startTime': startTime,
            'endTime': endTime,
            'paralell': False,
            'speakerId': speaker_info[i[1]['user_profile']],
            'content': []        
        }


        words = i[1]['text'].split(' ')
        for w_index, word in enumerate(words):
            phrase = {
                'type': 'phrase',
                'attrs': {
                    'startTime': startTime + w_index,
                    'endTime': startTime + (w_index + 1),
                    'type': 'WORD',
                    'confidence': 1
                },
                'content': [
                    {
                        'type': 'text',
                        'text': word
                    }
                ]
            }
            speaker_frame['content'].append(phrase)

        transcript['transcript']['content'][0]['content'].append(speaker_frame)
    
```

```python
import json
with open('nexoya - developer channel.json', 'w', encoding ='utf8') as json_file:
    json.dump(transcript, json_file, ensure_ascii = True)
```
